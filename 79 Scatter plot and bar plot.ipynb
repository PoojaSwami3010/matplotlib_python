# 79 Scatter plot and bar plo

### types:
   
   1.line
   2.scatter
   3.bar
   4.histogram
   5.subplots

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
# plot data and create line plot
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
# scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x)) # np.exp(x) is exponential value of x
```

```python
# another scatter plot
plt.scatter(x,np.sin(x))
```

```python
# bar plot using dictionary

dt={'apple':34,'banana':20,'pinaple':65,'grapes':24}
plt.bar(dt.keys(),dt.values(),color='m',width=0.5)
plt.title("Fruits vs prices")
plt.xlabel("fruits")
plt.ylabel("prices")
```

# 80 histogram and subplots

```python
# bar plot using dictionary And horizontal bar chart

dt={'apple':34,'banana':20,'pinaple':65,'grapes':24}
plt.barh(list(dt.keys()),list(dt.values()),color='y') # horizontal
plt.title("Fruits vs prices")
plt.xlabel("fruits")
plt.ylabel("prices")
```

```python
# histogram using fig and ax
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

```python
# histogram using plt directly
x=np.random.randn(1000)
plt.hist(x,color='y');# ';' removes array list look at above cell last line 
```

```python
# subplots option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(8,7))

# ploting data
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(['a','b','c'],[10,20,50])
ax4.hist(np.random.randn(1000));
```

# 81 subplot option 2

```python
# subplots option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,7))

# ploting data using index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(['a','b','c'],[10,20,50])
ax[1,1].hist(np.random.randn(1000));
```

```python
# another way

plt.subplot(2,2,1) # row,column,1 st position
plt.plot(x,x/2,'r')

plt.subplot(2,2,2) #  row,column,2nd position
plt.scatter(np.random.random(10),np.random.random(10),color='g')

plt.subplot(2,2,3) # row,column,3rd position
plt.hist(np.random.randn(1000),color='m')


plt.subplot(2,2,4) # row,column, 4th position
plt.bar(['a','b','c'],[10,20,30],color='y')
```
