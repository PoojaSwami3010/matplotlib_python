# 88 plotting pandas dataframe 5

 which on u should you use?  (pyplot vs matplotlib oo method)

 -when plotting somehing quickly, use pyplot
 - when plotting something advance ,use oo method
 

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
#len(over_50)
```

```python
# pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# OO method mixed with pyplot
fig,ax=plt.subplots(figsize=(8,5))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
ax.set_xlim([45,100])  # limitation of axis
```

# 89 plotting pandas dataframe 6

```python
# oo method from scratch
fig ,ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50['age'], y=over_50['chol'],c=over_50['target'])

# customised data
ax.set(title='heart disease',xlabel='age',ylabel='cholesterol')
#ax.legend(*scatter.legend_elements(),title='Target')

# horizontal line to mean of cholestroal
ax.axhline(over_50['chol'].mean(),linestyle='--')
```

# 90 plotting from pandas dataframe 7

```python
over_50.head()
```

```python
# subplot

fig,(ax0,ax1)=plt.subplots(ncols=1,nrows=2,figsize=(8,8))

# add data to ax0
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

# customize ax0
ax0.set(title="age vs chol",xlabel='age',ylabel='cholestrol')
# add a mean line
ax0.axhline(y=over_50['chol'].mean(),linestyle='--')

# add data to ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])

# customize ax1
ax1.set(title="age vs thal",xlabel='age',ylabel='cholestrol')
# add a mean line
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--')

# add title to figure
fig.suptitle("heart disease analysis",fontsize=16);
```

# 92 customize plot 2

```python
plt.style.use("seaborn")
```

```python
fig ,ax=plt.subplots(figsize=(10,6))

# plot the data and with color code plasma
scatter=ax.scatter(x=over_50['age'], y=over_50['chol'],c=over_50['target'],cmap='plasma')

# customised data
ax.set(title='heart disease',xlabel='age',ylabel='cholesterol')
#ax.legend(*scatter.legend_elements(),title='Target')

# horizontal line to mean of cholestroal
ax.axhline(over_50['chol'].mean(),linestyle='--')
```

```python
# subplot

fig,(ax0,ax1)=plt.subplots(ncols=1,nrows=2,figsize=(8,8))

# add data to ax0
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='plasma')

# customize ax0
ax0.set(title="age vs chol",xlabel='age',ylabel='cholestrol')
# add a mean line
ax0.axhline(y=over_50['chol'].mean(),linestyle='--')
ax0.set_xlim([50,80])
# add data to ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='plasma')

# customize ax1
ax1.set(title="age vs thal",xlabel='age',ylabel='thalach')
# add a mean line
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--')
ax1.set_xlim([50,80])
ax1.set_ylim([50,300])
# add title to figure
fig.suptitle("heart disease analysis",fontsize=16);
```
