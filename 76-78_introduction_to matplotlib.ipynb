### matplotlib workflow example

```python
# 0. import matplotlib and get ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2.setup plot 
fig,ax=plt.subplots(figsize=(8,8)) # (figure=(width,height))

# 3. plot data
plt.plot(x,y)

# 4.customise data
ax.set(title="sample_plot",
      xlabel="x_axis",
      ylabel="y-axis")

# 5. save figure and show
fig.savefig("sample_plot.png")
```
